```
import matplotlib.pyplot as plt # Подключаем модуль pyplot библиотеки matplotlib и задаем псевдоним plt
%matplotlib inline
```

```
import numpy as np # Подключаем библиотеку numpy
x = np.linspace(0,15,101) # Генерируем последовательность из 101 числа равнораспределенных в интервале от 0 до 15
print(x.shape) # Выводим размерность сформированного массива
print(x) # Выводим сам массив
```

#### Вывод графиков

```
plt.plot(x) # Визуализируем график
plt.show()  # Фиксируем вывод
```

```
y = np.cos(x) # Задаем значения, соответсвующие функции косинус
print(y) # Выводим сформированный массив y
plt.plot(y) # Визуализируем график
plt.show() # Фиксируем график
```

```
plt.plot(x, y)  # Визуализируем график, указав обе оси
plt.show() # Фиксируем график
```

Чем плох график выше?

* отсутствуют подписи осей
* отсутствует название графика

Давайте это поправим:

```
plt.plot(x, y) # Визуализируем график, указав обе оси
plt.title('cos(x)') # Задаем название графика
plt.xlabel('ось X') # Задаем название оси X
plt.ylabel('ось Y') # Задаем название оси Y
plt.show() # Фиксируем график
```

Добавим лейблы на график:

```
z = np.sin(x) / 2 # Задаем функцию sin(x)/2
plt.plot(x, y, label='cos(x)') # Визуализируем график для значения y (и задаем ему название (легенду))
plt.plot(x, z, label='sin(x)/2') # Визуализируем график для значения z (и задаем ему название (легенду))
plt.title('sin(x)/2 и cos(x)') # Задаем название графика
plt.xlabel('ось X') # ось х
plt.ylabel('Y, Z') # ось y,z
plt.legend() # Вызываем отображение легенды на графике
plt.show() # Фиксируем график
```

```
z = np.sin(x) / 2 # Задаем функцию sin(x)/2
plt.plot(x, y, label='cos(x)') # Визуализируем график для значения y (и задаем ему название (легенду))
plt.plot(x, z, label='sin(x)/2') # Визуализируем график для значения z (и задаем ему название (легенду))
plt.title('sin(x)/2 и cos(x)') # Задаем название графика
plt.xlabel('ось X') # ось х
plt.ylabel('Y, Z') # ось y,z
plt.legend(loc='upper left') # Вызываем отображение легенды на графике
plt.show() # Фиксируем график
```

```
# рисуем график
plt.figure(figsize=(15,6))
plt.plot(x, z, 'co') # для вывода голубыми кружками 'co'
plt.title('0.5 sin(x)')
plt.xlabel('х')
plt.ylabel('y')
plt.show()
```

Параметр linestyle:

```
styles = ['-', '--', '-.', ':', 'steps']

for ls in styles:
    # рисуем график
    plt.figure(figsize=(15,6))
    plt.plot(x, z, c='r', ls=ls) # т.к. в предыдущих примерах мы укзаывали цвет первой буквой и потом стиль линии,
                                 # то в этом случае, когда хотим менять только стиль линии, а цвет оставить тем же,
                                 # будем подставлять разные значения из списка стилей линий в параметр ls = 
    plt.title('0.5 sin(x)')
    plt.xlabel('х')
    plt.ylabel('y')
    plt.show()
```

Параметр marker:

```
styles = ['+', ',', '.', '1', '2', '3', '4' ]

for marker in styles:
    # рисуем график
    plt.figure(figsize=(15,6))
    plt.plot(x, z, c='r', marker=marker, ls=':', mec='b') # для параметра mec = указываем цвет маркера, а в marker, соответственно, тип маркера
    plt.title('0.5 sin(x)')
    plt.xlabel('х')
    plt.ylabel('y')
    plt.show()
```

Контроль толщины линии графика осуществляется за счет параметра linewidth:

```
# рисуем график
plt.figure(figsize=(15,6))
plt.plot(x, z, c='cornflowerblue', linewidth=2) # 'cornflowerblue'- цвет, linewidth = 2 - толщина 2 пикселя
plt.title('0.5 sin(x)')
plt.xlabel('х')
plt.ylabel('y')
plt.show()
```

## Несколько графиков в одних осях

```
plt.figure(1, figsize=(10,5)) # задаем размер полотна

# Альтернативная равносильная запись создания подграфиков 
plt.subplot(2, 1, 1) # указываем, что хотим создать дополнительный подграфик: 2 - количество строк, 1 - количество столбцов, 1 - какой по счету этот график
plt.title('Два графика с общей осью х')
plt.plot(x, y, 'b-.', label='cos(x)')
plt.legend()

plt.subplot(2, 1, 2) # указываем, что хотим создать дополнительный подграфик: 2 - количество строк, 1 - количество столбцов, 1 - какой по счету этот график
plt.plot(x, z, 'r--', label='0.5 sin(x)')
plt.xlabel('общая ось х')
plt.legend()

plt.show()
```

```
# Пример вывода трёх графиков
plt.figure(1, figsize=(10,5))

plt.subplot(221) # указываем, что хотим создать дополнительный подграфик: 2 - количество строк, 2 - количество столбцов, 1 - какой по счету этот график
plt.title('Три графика с общей осью х')
plt.plot(x, y, 'b-.', label='cos(x)')
plt.legend()

plt.subplot(222) # указываем, что хотим создать дополнительный подграфик: 2 - количество строк, 2 - количество столбцов, 2 - какой по счету этот график
plt.plot(x, z, 'g--', label='0.5 sin(x)')
plt.xlabel('общая ось х')
plt.legend()

plt.subplot(224) # указываем, что хотим создать дополнительный подграфик: 2 - количество строк, 2 - количество столбцов, 4 - какой по счету этот график
plt.plot(x, z, 'r--', label='0.5 sin(x)')
plt.xlabel('общая ось х')
plt.legend()

plt.show()
```

## Несколько графиков как один объект

*Разбор данного раздела:* https://youtu.be/lxHrkPMBwd0?t=1264


```
plt.subplots?
```

```
# Создаем игрушечные данные:
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)

# Задаем пространство для подграфиков
f, axes = plt.subplots(2, 2, sharex=False, sharey=False, figsize=(12,8)) # параметры shareх и sharey - если False, то у каждого графика
                                                                         # будет отображена своя ось х и y, если True, то оси х и у будут
                                                                         # общими для всего пространства графиков

# График с координатами положения (0,0) по отношению к оси 
axes[0,0].plot(x,y,label='sin(x^2)', c='r')
axes[0,0].legend() 

# График с координатами положения (0,1) по отношению к оси 
axes[0,1].plot(x, np.abs(y), label='np.abs(y)')
axes[0,1].legend()

# График с координатами положения (1,0) по отношению к оси 
axes[1,0].plot(x,y**3)

# График с координатами положения (1,1) по отношению к оси 
axes[1,1].plot(x,np.arccos(y))
plt.show()
```
