# Основы Комбинаторики

*Разбор данного раздела:*



```
# Реализация факториала через цикл
def factorial(n):
  f = 1
  for x in range(1, n+1):
    f *= x
  return f

# Перестановки без повторений - n!
def P(n):  
  return factorial(n)

# Сочетания без повторений - n! / ((n-m)! * m!)
def C_n_m(n, m):
  return factorial(n) / (factorial(n-m) * factorial(m))

# Размещения без повторений - C_n_m * P_m
def A_n_m(n, m):
  return C_n_m(n, m) * P(m)

# Перестановки с повторениями - n! / (n_1! * n_2! * n_3!...n_k!)
def P_rep(n, m):  # m принимается списком в котором хранятся числа повторений очередного объекта
  add = 1
  for x in m:
    add *= factorial(x)
  return factorial(n) / add

# Сочетания с повторениями - (n+m-1)! / ((n-1)! * m!)
def C_n_m_rep(n, m):
  return factorial(n+m-1) / (factorial(n-1) * factorial(m))

# Размещения с повторениями - n ** m
def A_n_m_rep(n, m):
  return n ** m
```

### Задача 1

Записать множество Е, если E = A ∪ B , причем А = {2, 4, 6, 8, 10, 12}, B = {3, 6, 9, 12}.

```
# Задаем два множества A и B как структуру set
A = {2, 4, 6, 8, 10, 12}
B = {3, 6, 9, 12}
```

```
# Объединение дает множество, содержащее в себе все элементы исходных множеств

# Способ 1 (оптимальный, через встроенный метод объединения множеств)
E = A.union(B)  # применяем метод объединения множеств A и B
print(E)  # выводим на экран
```

```
# Способ 2
E = A.copy()  # создаем копию множества A
E.update(B)  # делаем его 'обновление', добавляя элементы множ-ва B
print(E)  # выводим на экран
```

```
# Способ 3
E = A | B  # объединение через оператор '|'
print(E)  # выводим на экран
```

### Задача 2

Записать множество Е, если E = A \ B, причем А = {2, 4, 6}, B = {3, 6, 9}.


```
# Задаем два множества A и B как структуру set
A = {2, 4, 6}
B = {3, 6, 9}
```

```
# Разность двух множеств дает множество, в которое входят все элементы первого множества, не входящие во второе множество

# Способ 1 (оптимальный, через встроенный метод разности множеств)
E = A.difference(B)  # применяем метод вычисления разности множеств A и B
print(E)  # выводим на экран
```

```
# Способ 2
E = A - B  # привычный оператор разности '-' здесь также работает
print(E)  # выводим на экран
```

```
# Способ 3
E = {e for e in A if e not in B}  # создаем при помощи генератора множеств Python
print(E)  # выводим на экран
```

# Бинарная логика

*Разбор данного раздела:*





Под бинарной логикой подразумевается такая логика, которая предполагает только два исхода: либо да **True**, либо нет **False**. Рассмотрим основные логические операции:

* Конъюнкция - бинарная операция, логическое И, умножение, аналогично операции **min**, но для булевой логики. То есть результат равен наименьшему операнду.
> в Python обозначается как and

* Дизъюнкция - бинарная операция, логическое ИЛИ, сложение, аналогично операции **max**, но для булевой логики. То есть результат равен наибольшему операнду.
> в Python обозначается как or

* Отрицание - унарная операция (операция с одним входом), логическое НЕ, результат которой является противоположным первоначальному. 
> в Python обозначается как not 

Для чего нужно? 
> Для сравнения значений

Выражения, результатом которых является **True** или **False**
- $<$   меньше
- $>$  больше
- $<=$  меньше или равно
- $>=$  больше или равно
- $==$  равно
- $!=$  не равно


```
x = 5
y = 8

print("x == y:", x == y)
print("x != y:", x != y)
print("x < y:", x < y)
print("x > y:", x > y)
print("x <= y:", x <= y)
print("x >= y:", x >= y)
```

Представим такую ситуацию: по телевизору синоптик объявляет прогноз погоды на завтра и утверждает следующее:

если не будет ветра, то будет пасмурная погода без дождя;

если будет дождь, то будет пасмурно и без ветра;

если будет пасмурная погода, то будет дождь и не будет ветра.

Какая погода будет завтра?

```
# a = Ветер
# b = Пасмурно
# c = Дождь
for a in [False, True]:
  for b in [False, True]:
    for c in [False, True]:
      f1 = a or (b and (not c))
      f2 = (not c) or (b and (not a))
      f3 = (not b) or (c and (not a))
      f = f1 and f2 and f3
      print(a,b,c, ':', f)
```

# **Глоссарий**

**Перестановка (P)** характеризует, сколькими различными способами можно переупорядочить элементы множества, не меняя их количество.

*Пример*. Допустим, у вас есть четыре цифры и ваша задача нагенерировать всевозможные четырехзначные пароли из этих цифр.

**Сочетанием (C)** из n по k называется набор k элементов, выбранных из данного множества, содержащего n различных элементов. Не учитывает порядок элементов.

*Пример*. Дано пять ингредиентов, и нужно приготовить три коктейля. Независимо от того, добавите вы сначала клубнику, а потом арбуз, или же сначала арбуз, а потом клубнику, вкус коктейля будет тем же.

**Размещением (A)** (из n по k) называется упорядоченный набор из k различных элементов из некоторого множества различных n элементов. Учитывает порядок элементов.

*Пример*. Допустим, у вас десять цифр для генерации четырехзначного пароля. Даже если вы будете использовать те же четыре цифры из этих десяти, но в разной последовательности, пароли будут разными.
