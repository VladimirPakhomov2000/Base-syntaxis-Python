###**(!) Матричное умножение**

*Разбор данного раздела:*


Ячейка i, j новой матрицы равна сумме умножения по элементам
- строки i первой матрицы
- колонки j второй матрицы

Матричное_умножение.png

````
# Функция матричного умножения, принимающая на вход 2 матрицы
def prod_matrix(matrix1, matrix2):
  shape1 = matrix1.shape  # определим форму 1-ой матрицы
  shape2 = matrix2.shape  # определим форму 2-ой матрицы
  
  if shape1[1] != shape2[0]:  # если количество столбцов 1-ой матрицы не равно количеству строк 2-ой матрицы
    return np.array([])        # тогда умножение невозможно, возвращаем пустой массив
  
  out_matrix = np.zeros((shape1[0], shape2[1]))  # создадим матрицу из 0, формой (количество строк 1-ой)Х(количество столбцов 2-ой) матриц
  
  for i in range(shape1[0]):  # пройдемся по индексам строк 1-ой матрицы
    for j in range(shape2[1]):  # пройдемся по индексам столбцов 2-ой матрицы
      curr_cell = 0
      for t in range(shape1[1]):  # пройдемся по индексам стобцов 1-ой матрицы
        curr_cell += matrix1[i,t] * matrix2[t,j]  # посчитаем сумму произведения элементов строки 1-ой матрицы и столбцов 2-ой матрицы
      out_matrix[i,j] = curr_cell  # заменим значения в матрице состоящих из 0, на посчитанную сумму произведений
  
  return out_matrix  # вернем полученную матрицу
````

````
matrix1 = np.array([[0, 1, 2], [1, 2, 3], [2, 3, 4], [3, 4, 5]])  # определим матрицу 4х3
matrix2 = np.array([[0, 1, 2, 3], [1, 2, 3, 4], [2, 3, 4, 5]])  # определим матрицу 3х4

print(matrix1)
print('-' * 10)
print(matrix2)
````

###Транспонирование матриц

*Разбор данного раздела:* 




---

Транспонирование матрицы - это преобразование исходной матрицы к новому виду, в котором строки становятся столбцами, а столбцы строками.


Транспонирование.png

````
# Функция транспонирования матрицы, которая на вход принимает матрицы
def transpose_matrix(matrix):
  shape = matrix.shape  # определяем форму матрицы
  # определяем матрицу состоящую из 0, формой:
  # количество строк = количеству столбцов исходной матрицы
  # количество столбцов = количеству строк исходной матрицы 
  out_matrix = np.zeros((shape[1], shape[0]))
  for i in range(shape[0]):  # проходимся по всем индексам строк исходной матрицы
    for j in range(shape[1]):  # проходимся по всем индексам столбцов исходной матрицы
      out_matrix[j,i] = matrix[i,j]  # заменяем все значения столбцов нулевой матрицы значениями строк исходной матрицы
  
  return out_matrix
````

````
transposed_matrix = transpose_matrix(matrix1)  # транспонируем матрицу, используя написанную функцию

print('Оригинальная матрица\n', matrix1)
print()
print('Транспонированная матрица\n', transposed_matrix)
````

Обратная матрица, это такая матрица, при матричном умножении на которую, исходная матрица превратится в матрицу с 1 на диагонали и 0 в остальных ячейках(работает только для квадратных матриц)
![link text](https://drive.google.com/uc?id=11cxOXWSLI_LqEK2LnyW8F5fQWgSRRccN)

````
matrix_4_inv = np.array([[4, 1, 3], [3, 5, 2], [2, 6, 4]])  # определим матрицу 3х3
print('Исходная матрица\n', matrix_4_inv)  # исходная матрица
````

````
# Функция Python, которая выводит график y
# y - вектор значений функции "y = ..."
# a - начальная точка диапазона чисел для массива x
# b - конечная точка диапазона чисел для массива x
def plot_vect(y, a, b):
  x = np.arange(a, b)  # создаем массив значений по x
  # Отображаем график
  plt.plot(x, y)
  plt.show()

# Функция, которая выводит график Python функции f
def plot_func(f, a, b):
  x = np.arange(a, b)  # создаем массив значений по x
  # Вычисляем функцию в этих точках
  y = np.array([f(i) for i in x])
  
  plot_vect(y, a, b)  # рисуем график

  return y   # возвращаем массив значений y
````

````
# Пример функции для графика
def f(x):
  if x < 0:  # для всех отрицательных иксов применяем предыдущую формулу x*x/1000 + x + 12
    y = x*x/1000 + x + 12
  else:  # если же x >= 0, то меняем формулу на x/1000 + x + 12
    y = x/1000 + x + 12

  return y   # возвращаем y
````

````
# Выводим график функции
y = plot_func(f, -2000, 1000)
````

````
# Вывести график функции с экстремумами
def plot_func_with_extremums(f, a, b):
  x = np.arange(a, b)  # создаем массив значений по x
  y = np.array([f(i) for i in x])  # вычисляем функцию в этих точках

  # Пустые листы для экстремумов
  local_mins = []  # локальные минимумы
  local_maxs = []  # локальные максимумы
  global_mins = []  # глобальный минимум
  global_maxs = []  # глобальный максимум

  # Минимальное и максимальное значение y
  min_y = np.min(y)
  max_y = np.max(y)

  # Проходим по диапазону массива значений функции
  for i in range(len(y)):
    if (y[i] == min_y):  # если точка равна минимуму y
      global_mins.append(i)  # то это глобальный минимум, добавляем в global_mins
    
    if (y[i] == max_y):  # если точка равна максимуму y
      global_maxs.append(i)  # то это глобальный максимум, добавляем в global_maxs

    # Если это не первая и не последняя точка...
    # ...(в них не можем считать локальные минимумы и максимумы)
    if (i > 0 and i < len(y)-1):     
      if (y[i-1] > y[i] and y[i+1] > y[i]):  # если точка меньше, чем две соседние
        local_mins.append(i)  # то это локальный минимум, добавляем в local_mins

      if (y[i-1] < y[i] and y[i+1] < y[i]):  # если точка больше, чем две соседние
        local_maxs.append(i)  # то это локальный максимум, добавляем в local_maxs

  # Выводим график функции y
  plt.plot(x, y)

  # Выводим глобальный минимум
  plt.plot([x[i] for i in global_mins],
            [y[i] for i in global_mins],
            'gs')
  
  # Выводим глобальный максимум
  plt.plot([x[i] for i in global_maxs],
            [y[i] for i in global_maxs],
            'rs')
  
  # Выводим локальные минимумы
  plt.plot([x[i] for i in local_mins],
            [y[i] for i in local_mins],
            'co')
  
  # Выводим локальные максимумы
  plt.plot([x[i] for i in local_maxs],
            [y[i] for i in local_maxs],
            'yo')
````

````
#Выводим функцию с экстремумами
plt.figure(figsize=(10,10))
plot_func_with_extremums(f4, -1000, 1000)
````

````
# Подключаем библиотеки для отображения 2d поверхностей
import pylab
from mpl_toolkits.mplot3d import Axes3D
````

````
# Функция отображения 2d поверхности
def plot_2d_func1():
  # Формируем точки по двум осям
  x = np.arange(-10, 10, 0.1)
  y = np.arange(-10, 10, 0.1)
  # Формируем двумерную сетку
  xgrid, ygrid = np.meshgrid(x, y)

  # Вычисляем 2d функцию
  zgrid = np.sin(xgrid) * np.sin(ygrid) / (xgrid * ygrid)
  
  # Создаём фигуру и 3D оси
  fig = pylab.figure()
  axes = Axes3D(fig)

  # Отображаем двумерную функцию
  axes.plot_surface(xgrid, ygrid, zgrid)
  pylab.show()
````

````
# Выводим двумерную функцию
plot_2d_func1()
````

````
# Аналогичная функция с другой 2d поверхностью
def plot_2d_func2():
  x = np.arange(-10, 10, 0.1)
  y = np.arange(-10, 10, 0.1)
  xgrid, ygrid = np.meshgrid(x, y)

  zgrid = (xgrid ** 2) * ygrid
  
  fig = pylab.figure()
  axes = Axes3D(fig)

  axes.plot_surface(xgrid, ygrid, zgrid)

  pylab.show()
````

````
# Отображаем 2d поверхность
plot_2d_func2()
````

````
# Функция вычисления производной функции
def get_derivative(f, a, b):
  x = np.arange(a,b)  # создаем массив значений по x

  # Рассчитываем производную
  dy = np.array([f(x[i+1]) - f(x[i]) for i in range(x.size-1)])

  return dy  # возвращаем производную
````

**Вторая производная**

*Разбор данного раздела:* 

````
# Функция расчёта второй производной
def get_second_derivative(f, a, b):
  x = np.arange(a,b)  # создаем массив значений по x

  # Вычисляем вторую производную
  ddy = np.array([f(x[i+2]) - 2*f(x[i+1]) + f(x[i]) for i in range(x.size-2)])
  
  return ddy  # возвращаем вторую производную
````

````
# Выводим график функции
y1 = plot_func(f1, -1000, 1000)
dy = get_derivative(f1, -1000, 1000)
# Вычисляем вторую производную и выводим график
ddy = get_second_derivative(f1, -1000, 1000)
plot_vect(dy, -999, 1000)
plot_vect(ddy, -999, 999)
````

# **Глоссарий**

**Скаляр** — величина, полностью определяемая в любой координатной системе одним числом или функцией, которое не меняется при изменении пространственной системы координат.

*Пример*: scalar = 2.0



**Вектор** - математический объект, характеризующийся величиной и направлением.

*Пример*: vector = array([1,2]) 
- размер (2,1)


**Матрицы** - математический объект, записываемый в виде прямоугольной таблицы целых, действительных или комплексных чисел, которая представляет собой совокупность строк и столбцов, на пересечении которых находятся её элементы. Количество строк и столбцов задает размер матрицы.

*Пример*: matrix = array([[1,2],[2,3],[5,6]])
 - размер (3,2)

```
np.dot(matrix1,matrix2) - матричное умножение

np.linalg.inv(matrix) - найти обратную матрицу

np.transpose(matrix1) - транспонировать матрицу
```



**Функция** – это зависимость одной переменной величины от другой. 

**Первая производная** характеризует скорость изменения функции в данной точке

**Вторая производная** характеризует мгновенное ускорение в заданной точке, а также выпуклость и вогнутость. 

**Экстремум функции** - максимальное/минимальное значение функции 


>Соответственно, если первая производная *дважды* дифференцируемой функции равна *нулю* в некоторой точке , а вторая производная в этой точке *положительна* (*отрицательна*), то есть точка *минимума* (*максимума*) функции.
